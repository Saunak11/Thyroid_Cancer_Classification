```python
#separation into benign and malignant before augmentation

import os
import shutil
import pandas as pd

# Load the matched images file
matched_csv_path = "matched_images.csv"
df_matched = pd.read_csv(matched_csv_path)

# Load the original balanced CSV file with classifications
balanced_csv_path = "balanced_tirads_data.csv"
df_balanced = pd.read_csv(balanced_csv_path)

# Define source image folder
image_folder = "image files"  # Update this to your actual image folder path

# Define destination folders
benign_folder = "benign"
malignant_folder = "malignant"

# Create folders if they don't exist
os.makedirs(benign_folder, exist_ok=True)
os.makedirs(malignant_folder, exist_ok=True)

# Create a dictionary mapping file names to classifications
classification_dict = dict(zip(df_balanced["File Name"].astype(str), df_balanced["classification"]))

# Move images based on classification
for _, row in df_matched.iterrows():
    file_number = str(row["File Name"])  # Get the file number
    images = eval(row["Matched Images"])  # Convert string list to actual list
    
    # Get classification (benign/malignant)
    classification = classification_dict.get(file_number, "Unknown")
    
    # Move images to corresponding folder
    for image in images:
        src_path = os.path.join(image_folder, image)
        
        # Check if file exists before moving
        if not os.path.exists(src_path):
            print(f"Skipping {image} - File not found!")
            continue  # Skip this file
        
        if classification == "benign":
            dest_path = os.path.join(benign_folder, image)
        elif classification == "malignant":
            dest_path = os.path.join(malignant_folder, image)
        else:
            continue  # Skip if classification is unknown
        
        shutil.move(src_path, dest_path)
        print(f"Moved {image} to {classification} folder")

print("Image separation completed!")
```

```python
import os
import pandas as pd
import shutil
import ast  # To convert string list to Python list

# Define paths
image_folder = "image files"  # Folder containing remaining 87 images
benign_folder = "benign"      # Destination folder for benign images
malignant_folder = "malignant"  # Destination folder for malignant images
csv_path = "balanced_tirads_data.csv"  # CSV with classification
matched_csv_path = "matched_images.csv"  # CSV with matched images

# Load CSV files
df_classification = pd.read_csv(csv_path)  # Contains benign/malignant classification
df_matched = pd.read_csv(matched_csv_path)  # Contains image mappings

# Convert 'Matched Images' column from string to list
df_matched['Matched Images'] = df_matched['Matched Images'].apply(ast.literal_eval)

# Create a dictionary mapping filenames to their corresponding images
file_to_images = dict(zip(df_matched['File Name'].astype(str), df_matched['Matched Images']))

# Extract all image filenames in 'image files' folder
actual_images = set(os.listdir(image_folder))

# Identify 87 images that have not been moved yet
remaining_images = []
for file_num, images in file_to_images.items():
    for img in images:
        if img in actual_images:
            remaining_images.append((file_num, img))

# Create a dictionary for classification (benign/malignant)
classification_dict = dict(zip(df_classification['File Name'].astype(str), df_classification['classification']))  # Assuming 'Label' column has 'benign'/'malignant'

# Move images based on classification
for file_num, img in remaining_images:
    src_path = os.path.join(image_folder, img)
    
    # Determine destination folder
    label = classification_dict.get(file_num, "unknown")  # Default to 'unknown' if not found
    if label == "benign":
        dest_folder = benign_folder
    elif label == "malignant":
        dest_folder = malignant_folder
    else:
        print(f"Skipping {img}: Classification not found!")
        continue

    dest_path = os.path.join(dest_folder, img)

    # Move the image
    shutil.move(src_path, dest_path)
    print(f"Moved {img} → {dest_folder}")

print(f"Total {len(remaining_images)} images moved successfully!")
```
